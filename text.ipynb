## General Quantization

```python
import torch
import torch.nn as nn

def linear_quantize(tensor, scale, zero_point, dtype=torch.int8):
    # symmetric (zero_point=0) 기준, clamp 후 round
    scaled = tensor /scale
    rounded = torch.round(scaled)
    clamped = torch.clamp(rounded, -128, 127) 
    return clamped.to(dtype)

def linear_dequantize(q_tensor, scale, zero_point=0):
    dequant = (q_tensor.float() - zero_point) * scale
    return dequant
```
